```python
def decRisk(x):
    if x < 200000:
        return ''
    elif (x >= 200000) & (x < 250000):
        return ''
    elif (x >= 250000) & (x < 300000):
        return ''
    elif x > 300000:
        return ''
```

```python
url = "https://api.upbit.com/v1/candles/minutes/1"
```

```python
querystring = {"market" : "KRW-DOGE","count":"200"}
headers = {"Acppet" : "application/json"}

response = requests.request("GET", url, headers=headers, params=querystring)
```

```python
res = response.json()
```

```python
res
```

```python
bit_table = pd.DataFrame(res)
```

# candle_data_time_kst    .
# high_price - low_price  
#  : diff_price

```python
bit_table['candle_date_time_kst'] = bit_table['candle_date_time_kst'].str.replace('2023-03-08T','')
```

```python
for i in range(len(bit_table)):
    st = bit_table.iloc[i].candle_date_time_utc
    st2 = st.split('T')[-1]
    bit_table['candle_date_time_utc'].lioc[i] = st2
```

```python
bit_table['diff_price'] = 'none'
```

```python
bit_table['diff_price'] = (bit_table.high_price - bit_table.low_price)
```
