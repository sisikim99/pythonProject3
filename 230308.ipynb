# table이라는 리스트안에 1,2,3번째 인덱스를 한 묶음으로
# 4,5,6 번째 인덱스를 한 묶음으로 해서 이중 리스트를 만들고 데이터 프레임까지 생성

```python
tmp =  []
for i in range(len(table[3].select('th'))):
    tmpList = []
    tmpList.append(table[0].select('th')[i].text)
    tmpList.append(table[1].select('td')[i].text)
    tmpList.append(table[2].select('td')[i].text)
    tmp.append(tmpList)
    
    tmpList = []
    tmpList.append(table[3].select('th')[i].text)
    tmpList.append(table[4].select('td')[i].text)
    tmpList.append(table[5].select('td')[i].text)
    tmp.append(tmpList)
```

```python
tmp
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(tmp,columns = ['지역','확진자 수','전날 대비 변화량'])
```

```python
df
```

```python
df['확진자 수'] = df['확진자 수'].str.replace(',','').astype('int')
```

```python
df
```

```python
df[df['확진자 수'] > 250000]
```

# 새로운 컬럼 : 위험도
# 주의 : 15만 미만
# 위험 : 20만~25만 미만
# 경고 : 25만~30만 미만
# 최고 : 30만 초과

```python
df['위험도'] = 'none'
```

```python
def decRisk(x):
    if x < 200000:
        return '주의'
    elif (x >= 200000) & (x < 250000):
        return '위험'
    elif (x >= 250000) & (x < 300000):
        return '경고'
    elif x > 300000:
        return '최고'
```

```python
df['위험도'] = df.apply(lambda x : decRisk(x['확진자 수']), axis = 1)
```

```python
df
```

```python
df.groupby('위험도').mean().round(2)
```

```python
df.groupby('위험도')[['확진자 수']].max()
```

```python
df.groupby('위험도')[['확진자 수']].min()
```

```python
df.sort_values(by = ['확진자 수'], ascending = False)
```

# candle_data_time_kst의 날짜를 없애고 시간만 남기기.
# high_price - low_price 결과값 계산
# 컬럼명 : diff_price

```python
bit_table['candle_date_time_kst'] = bit_table['candle_date_time_kst'].str.replace('2023-03-08T','')
```

```python
for i in range(len(bit_table)):
    st = bit_table.iloc[i].candle_date_time_utc
    st2 = st.split('T')[-1]
    bit_table['candle_date_time_utc'].lioc[i] = st2
```

```python
bit_table['diff_price'] = 'none'
```

```python
bit_table['diff_price'] = (bit_table.high_price - bit_table.low_price)
```
