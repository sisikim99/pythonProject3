# KDA
# (K+A)/D
# KDA컬럼을 추가
# none -> apply를 사용해서 kda 계산 소수점 둘째자리까지 사용
# KDA가 높은게 위로 가도록 sort

```python
df['KDA'] = 'none'
```

```python
df
```

```python
def KDA(x):
    if x.deaths == 0:
        x.deaths = 1
        return round((x.kills+x.assists)/x.deaths,2)
    else :
        return round((x.kills+x.assists)/x.deaths,2)
```

```python
df['KDA'] = df.apply(lambda x : KDA(x) , axis = 1)
```

```python
df
```

```python
df.sort_values(by = 'KDA', ascending = False)
```

```python
damage = data_solo_rank[['damageDealt', 'damageTaken', 'kills', 'deaths',
       'assists']]
```

```python
damage['KDA'] = 'none'
```

```python
damage
```

```python
def KDA2(x):
    if x.deaths == 0:
        x.deaths = 1
        return round((x.kills+x.assists)/x.deaths,2)
    else :
        return round((x.kills+x.assists)/x.deaths,2)
```

```python
damage['KDA'] = damage.apply(lambda x : KDA2(x) , axis = 1)
```

```python
damage
```

```python
damage['Damage'] = 'none'
```

```python
def dam(x):
    if x.damageTaken == 0:
        x.damageTaken = 1
        return x.damageDealt/x.damageTaken
    else :
        return x.damageDealt/x.damageTaken
```

```python
damage['Damage'] = damage.apply(lambda x : dam(x) , axis = 1)
```
