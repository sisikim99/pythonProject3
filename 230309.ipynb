# 전월세 = 전세/월세 - 전세,월세 각각 뽑아서 데이터프레임 만들기
# astype -> 보증금과 임대료를 astype으로 int,float로 변경
# 분리된 데이터프레임으로 자치구명을 기준으로 groupby 해서 보증금의 평균 계산
# groupby된 데이터 프레임을 sort해서 내림차순으로 정렬

# 풀이

```python
tmp = my_df[my_df['전월세'] == '전세']
tmp2 = my_df[my_df['전월세'] == '월세']
```

```python
tmp = tmp.astype({'보증금':'float','임대료':'float'})
tmp2 = tmp.astype({'보증금':'float','임대료':'float'})
```

```python
groupDf = tmp.groupby('자치구명')[['보증금','임대료']].mean().round(2)
```

```python
groupDf2 = tmp2.groupby('자치구명')[['보증금','임대료']].mean().round(2)
```

```python
groupDf.sort_values(by=['보증금'],ascending = False)
```

```python
groupDf2.sort_values(by=['임대료'],ascending = False)
```

```python
groupDf2['임대료/보증금'] = (groupDf2['임대료']/groupDf2['보증금']).round(4)
```

```python
result_df = groupDf2.sort_values(by='임대료/보증금')
```
